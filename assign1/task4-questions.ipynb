### Question 2 
What are the differences between stochastic gradient descent and Batch gradient descent? Why is mini-batch gradient descent used in practice?

   Your answer: **[1.The difference between these are the amount of data used for update. Batch gradient uses the entire training set and passes all training samples simultaneously in a large batch. SGD uses only a single or a subset of examples at a time and this may slow down the calculation. Therefore, batch gradient descent converges to the minimum with small fluctuation, SGD could jump to new local minima, and converges faster for larger datasets.<br>
2.Mini-batch is a mixture of Batch Gradient Descent and SGD. It reduces the variance of parameter updates comparing to SGD, and larger batches provides a more accurate estimate for gradient, make use of matrix optimizations, and small batches offering a regularizing effect. 
]**

   Your answer: **[1.The activation function is an abstraction representing the rate of action potential firing in the cell. It helps the network to learn complex patterns in data and can introduce non-linearity into the output of a neuron. <br>
2.Vanishing gradients can arise when using sigmoid or tanh activation functions. As backpropagation computes gradients by the chain rule and the gradient of sigmoid and tanh becomes increasingly small as the absolute value of x increases, this has the effect of multiplying n of these small numbers to compute gradients of the early layers in an n-layer network, meaning that the gradient (error signal) decreases exponentially with n while the early layers train very slowly.]**
